```python
# câu 5.1
import sqlite3

def connect_sqlite(db_name):
    try:
        # Kết nối (tạo DB nếu chưa tồn tại)
        conn = sqlite3.connect(db_name)
        print("Kết nối SQLite thành công")

        # Tạo con trỏ
        cursor = conn.cursor()

        # Lấy version SQLite
        cursor.execute("SELECT sqlite_version();")
        record = cursor.fetchone()

        print("Phiên bản SQLite là:", record[0])

        # Đóng kết nối
        cursor.close()
        conn.close()
        print("Đã đóng kết nối SQLite")

    except sqlite3.Error as error:
        print("Lỗi khi kết nối SQLite:", error)

# Gọi hàm
connect_sqlite("test_database.db")
```

```python
# câu 5.3
import sqlite3

# Kết nối (tạo CSDL nếu chưa tồn tại)
conn = sqlite3.connect("array_db.db")
cursor = conn.cursor()

print("Kết nối SQLite thành công")

# Tạo bảng (đóng vai trò như mảng)
cursor.execute("""
CREATE TABLE IF NOT EXISTS numbers (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    value INTEGER
)
""")
print("Đã tạo bảng numbers")

# Dữ liệu mảng trong Python
array_data = [10, 20, 30, 40, 50]

# Chèn các phần tử mảng vào CSDL
for item in array_data:
    cursor.execute("INSERT INTO numbers (value) VALUES (?)", (item,))

conn.commit()
print("Đã lưu mảng vào cơ sở dữ liệu")

# Lấy dữ liệu ra kiểm tra
cursor.execute("SELECT value FROM numbers")
rows = cursor.fetchall()

print("Dữ liệu trong bảng (mảng):")
for row in rows:
    print(row[0])

# Đóng kết nối
cursor.close()
conn.close()
print("Đã đóng kết nối SQLite")
```

```python
# câu 5.4
import sqlite3

def list_tables(db_file):
    try:
        # Kết nối tới tệp CSDL SQLite
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()
        print(f"Kết nối thành công tới cơ sở dữ liệu: {db_file}")

        # Truy vấn danh sách các bảng
        cursor.execute("""
        SELECT name FROM sqlite_master
        WHERE type='table'
        ORDER BY name;
        """)

        tables = cursor.fetchall()

        if tables:
            print("Các bảng trong cơ sở dữ liệu:")
            for table in tables:
                print("-", table[0])
        else:
            print("Cơ sở dữ liệu không có bảng nào")

        # Đóng kết nối
        cursor.close()
        conn.close()

    except sqlite3.Error as error:
        print("Lỗi SQLite:", error)

# Gọi hàm (thay tên file CSDL nếu cần)
list_tables("array_db.db")
```

```python
# câu 5.5
import sqlite3

# Kết nối tới cơ sở dữ liệu (tạo nếu chưa có)
conn = sqlite3.connect("example.db")
cursor = conn.cursor()

print("Kết nối SQLite thành công")

# 1. Tạo bảng
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    score REAL
)
""")
print("Đã tạo bảng students")

# 2. Chèn dữ liệu vào bảng
students_data = [
    ("An", 20, 8.5),
    ("Bình", 21, 7.8),
    ("Chi", 19, 9.2)
]

cursor.executemany(
    "INSERT INTO students (name, age, score) VALUES (?, ?, ?)",
    students_data
)
conn.commit()
print("Đã chèn dữ liệu vào bảng")

# 3. Lấy tất cả các bản ghi
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

# 4. Hiển thị kết quả
print("\nDanh sách sinh viên:")
for row in rows:
    print(f"ID: {row[0]}, Tên: {row[1]}, Tuổi: {row[2]}, Điểm: {row[3]}")

# Đóng kết nối
cursor.close()
conn.close()
print("\nĐã đóng kết nối SQLite")
```

```python
# câu 5.10
import sqlite3

DB_NAME = "product.db"

# Kết nối CSDL
def connect_db():
    return sqlite3.connect(DB_NAME)

# Tạo bảng product
def create_table():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS product (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        price REAL NOT NULL,
        amount INTEGER NOT NULL
    )
    """)
    conn.commit()
    conn.close()

# Hiển thị sản phẩm
def show_products():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM product")
    rows = cursor.fetchall()

    print("\nDanh sách sản phẩm:")
    for row in rows:
        print(row)

    conn.close()

# Thêm sản phẩm
def add_product():
    name = input("Tên sản phẩm: ")
    price = float(input("Giá: "))
    amount = int(input("Số lượng: "))

    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO product (name, price, amount) VALUES (?, ?, ?)",
        (name, price, amount)
    )
    conn.commit()
    conn.close()
    print("Đã thêm sản phẩm")

# Tìm sản phẩm theo tên
def search_product():
    keyword = input("Nhập tên sản phẩm cần tìm: ")

    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute(
        "SELECT * FROM product WHERE name LIKE ?",
        ('%' + keyword + '%',)
    )
    rows = cursor.fetchall()

    print("\nKết quả tìm kiếm:")
    for row in rows:
        print(row)

    conn.close()

# Cập nhật sản phẩm
def update_product():
    pid = int(input("Nhập ID sản phẩm: "))
    price = float(input("Giá mới: "))
    amount = int(input("Số lượng mới: "))

    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute(
        "UPDATE product SET price=?, amount=? WHERE id=?",
        (price, amount, pid)
    )
    conn.commit()
    conn.close()
    print("Đã cập nhật sản phẩm")

# Xoá sản phẩm
def delete_product():
    pid = int(input("Nhập ID sản phẩm cần xoá: "))

    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM product WHERE id=?", (pid,))
    conn.commit()
    conn.close()
    print("Đã xoá sản phẩm")

# Menu chính
def menu():
    create_table()
    while True:
        print("""
====== QUẢN LÝ SẢN PHẨM ======
1. Hiển thị sản phẩm
2. Thêm sản phẩm
3. Tìm sản phẩm theo tên
4. Cập nhật sản phẩm
5. Xoá sản phẩm
0. Thoát
""")
        choice = input("Chọn chức năng: ")

        if choice == "1":
            show_products()
        elif choice == "2":
            add_product()
        elif choice == "3":
            search_product()
        elif choice == "4":
            update_product()
        elif choice == "5":
            delete_product()
        elif choice == "0":
            break
        else:
            print("Lựa chọn không hợp lệ!")

# Chạy chương trình
menu()
```

```python
# câu 5.11
import sqlite3

DB_NAME = "ql_nhan_vien.db"

def connect_db():
    return sqlite3.connect(DB_NAME)

# Tạo bảng
def create_tables():
    conn = connect_db()
    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS PHONG (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        ten_phong TEXT NOT NULL
    )
    """)

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS NHAN_VIEN (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        ho_ten TEXT NOT NULL,
        tuoi INTEGER NOT NULL,
        dia_chi TEXT NOT NULL,
        luong REAL NOT NULL,
        id_phong INTEGER,
        FOREIGN KEY (id_phong) REFERENCES PHONG(id)
    )
    """)

    conn.commit()
    conn.close()

# Thêm phòng ban
def add_phong():
    ten = input("Tên phòng: ")
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("INSERT INTO PHONG (ten_phong) VALUES (?)", (ten,))
    conn.commit()
    conn.close()
    print("Đã thêm phòng")

# Hiển thị phòng ban
def show_phong():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM PHONG")
    for row in cursor.fetchall():
        print(row)
    conn.close()

# Thêm nhân viên
def add_nhan_vien():
    ho_ten = input("Họ tên: ")
    tuoi = int(input("Tuổi: "))
    dia_chi = input("Địa chỉ: ")
    luong = float(input("Lương: "))
    id_phong = int(input("ID phòng: "))

    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("""
    INSERT INTO NHAN_VIEN (ho_ten, tuoi, dia_chi, luong, id_phong)
    VALUES (?, ?, ?, ?, ?)
    """, (ho_ten, tuoi, dia_chi, luong, id_phong))
    conn.commit()
    conn.close()
    print("Đã thêm nhân viên")

# Hiển thị nhân viên
def show_nhan_vien():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("""
    SELECT NV.id, NV.ho_ten, NV.tuoi, NV.dia_chi, NV.luong, P.ten_phong
    FROM NHAN_VIEN NV
    JOIN PHONG P ON NV.id_phong = P.id
    """)
    for row in cursor.fetchall():
        print(row)
    conn.close()

# Menu
def menu():
    create_tables()
    while True:
        print("""
====== QUẢN LÝ NHÂN VIÊN ======
1. Thêm phòng ban
2. Hiển thị phòng ban
3. Thêm nhân viên
4. Hiển thị nhân viên
0. Thoát
""")
        ch = input("Chọn chức năng: ")

        if ch == "1":
            add_phong()
        elif ch == "2":
            show_phong()
        elif ch == "3":
            add_nhan_vien()
        elif ch == "4":
            show_nhan_vien()
        elif ch == "0":
            break
        else:
            print("Lựa chọn không hợp lệ")

menu()
```
