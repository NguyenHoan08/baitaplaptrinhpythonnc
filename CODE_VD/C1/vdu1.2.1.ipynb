```python
class Dog:

    def __init__(self,name,size,age,color):
        self.__name=name
        self.__size=size
        self.__age=age
        self.__color=color
    @property
    def age(self):
        print("Getter for age called")
        if not hasattr(self,"_Dog__age"):
            raise AttributeError("Age has been seleted or not set")
        return self.__age 
    @age.setter 
    def age(self,value):
        print("Setter for age called ")
        if value<0:
            raise ValueError("Age cannot be negative")
        self.__age=value
    @age.deleter 
    def age(self):
        print("Deleter for age called")
        if not hasattr(self,"_Dog__age"):
            raise AttributeError("Age has already been deleted or not set")
        del self.__age

dog=Dog("Bull","large",2,"yellow")
print(f"Bull's age is :(dog.age)")
dog.age=3
print(f"Bull updated Age is:{dog.age}")
```

```python
class LopCha:
    def __init__(self, ten_cha):
        self.ten_cha = ten_cha 
    def thong_tin(self):
        print("Tôi là Lớp Cha, tên của tôi là", self.ten_cha)
class LopCon(LopCha):
    def __init__(self, ten_cha, ten_con):
        super().__init__(ten_cha) 
        self.ten_con = ten_con
    def thong_tin(self):
        print("Tôi là Lớp Con, tên của tôi là", self.ten_con)
        super().thong_tin()
class LopChau(LopCon):
    def __init__(self, ten_cha, ten_con, ten_chau):
        super().__init__(ten_cha, ten_con) 
        self.ten_chau = ten_chau
    def thong_tin(self):
        print("Tôi là Lớp Cháu, tên của tôi là", self.ten_chau) 
        super().thong_tin()

lop_chau = LopChau("Ông A", "Bác B", "Cháu C") 
lop_chau.thong_tin()
```
