```python
# Khai báo lớp cơ sở (Lớp cha)
class SinhVien:
    def __init__(self, ma, hoten, lop):
        self.ma = ma
        self.hoten = hoten
        self.lop = lop

    def info(self):
        return f"Mã SV: {self.ma} | Họ tên: {self.hoten} | Lớp: {self.lop}"

#  Định nghĩa lớp svk18a3 kế thừa từ lớp SinhVien
class svk18a3(SinhVien):
    def __init__(self, ma, hoten, lop, chuyen_nganh):
        # Sử dụng super() để gọi hàm khởi tạo của lớp cha
        super().__init__(ma, hoten, lop)
        self.chuyen_nganh = chuyen_nganh

    #  Ghi đè phương thức info() (Method Overriding)
    def info(self):
        # Gọi lại thông tin từ lớp cha và cộng thêm thông tin riêng của lớp con
        base_info = super().info()
        return f"{base_info} | Chuyên ngành: {self.chuyen_nganh} (Lớp chuyên biệt K18A3)"

#  Khai báo đối tượng và kiểm tra tính Đa hình
if __name__ == "__main__":
    # Khai báo một đối tượng của lớp con
    sv = svk18a3("SV001", "Nguyễn Văn A", "K18A3", "Công nghệ phần mềm")

    # Gọi phương thức info()
    # Mặc dù sv kế thừa SinhVien, nhưng do ta đã ghi đè, 
    # Python sẽ ưu tiên gọi phương thức của lớp svk18a3.
    print(sv.info())
```
